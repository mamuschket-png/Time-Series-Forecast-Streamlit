# LSTM Cheatsheet

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense

# Load the Airline Passengers dataset
url = "https://raw.githubusercontent.com/jbrownlee/Datasets/master/airline-passengers.csv"
df = pd.read_csv(url)

# Prepare the data
data = df['Passengers'].values.astype('float32')
```

```python
data
```

```python
# Normalize the data
scaler = MinMaxScaler(feature_range=(0, 1))
data = scaler.fit_transform(data.reshape(-1, 1))
```

```python
data
```

```python
data
```

```python
# Modified create_sequences function
def create_sequences(data, input_steps, output_steps):
    """
    Create sequences for time series prediction

    Parameters:
    data: input time series data
    input_steps: number of time steps in the input sequence
    output_steps: number of time steps in the output sequence
    """
    X, y = [], []
    for i in range(len(data) - input_steps - output_steps + 1):
        # Input sequence (features)
        X.append(data[i:(i + input_steps), 0])
        # Output sequence (targets)
        y.append(data[(i + input_steps):(i + input_steps + output_steps), 0])
    return np.array(X), np.array(y)
```

```python
# Create sequences: 12 months input -> 3 months output
input_steps = 12
output_steps = 7
X, y = create_sequences(data, input_steps, output_steps)
```

```python
y
```

```python
X = X.reshape((X.shape[0], X.shape[1], 1))
```

```python
# SUPER SIMPLE MODEL -
model = Sequential([
    LSTM(5, input_shape=(input_steps, 1)),  # LSTM layer

    Dense(output_steps)                       # Single output layer
])

model.compile(optimizer='adam', loss='mse')
model.summary()

# Train
model.fit(X, y)
```

```python
model.predict(data[-input_steps:].reshape(1, input_steps, 1))
```

```python

# Make prediction
last_sequence = data[-input_steps:].reshape(1, input_steps, 1)
prediction = model.predict(last_sequence)

print(f"Next {output_steps} days predictions: {prediction[0]}")
```

## Iterative future predictor

```python
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense

# Assuming you already have:
# - trained model
# - scaler (if you scaled the data)
# - input_steps = 12
# - data (your original or scaled data)

def predict_future(model, last_sequence, n_future):
    """
    Predict future values iteratively

    Parameters:
    model: trained LSTM model
    last_sequence: last known sequence of data (length = input_steps)
    n_future: how many future steps to predict
    """
    future_predictions = []
    current_sequence = last_sequence.copy()

    for i in range(n_future):
        # Make prediction for next step
        next_pred = model.predict(current_sequence.reshape(1, input_steps, 1), verbose=0)
        future_predictions.append(next_pred[0, 0])

        # Update the sequence: remove first element, add new prediction
        current_sequence = np.roll(current_sequence, -1)
        current_sequence[-1] = next_pred[0, 0]

    return np.array(future_predictions)

# Example usage:
input_steps = 12
n_future = 6  # Predict next 6 time steps

# Get last known sequence from your data
last_sequence = data[-input_steps:]  # Last 12 known values

# Get future predictions
future_predictions = predict_future(model, last_sequence, n_future)

print(f"Next {n_future} predictions: {future_predictions}")
```
