```python
# (1) Calculate z-scores to identify outliers
# => Bei einem Z-Score > 3 (bzw. < -3) spricht man typischerweise von Ausreißern

from scipy import stats
import numpy as np

z_scores = stats.zscore(df['unit_sales'])
z_scores_BETRAG = np.abs(z_scores)
outliers = df[z_scores_BETRAG > 3]
outliers
```

## Seasonal Decomposition
### Theorie



- Generell scheint zu gelten: Man kann ja auch muliplikativ ausprobieren und dann die Residuen mit der additativen Variante vgl. => umso kleiner desto besser
- Entscheidung ob additativ oder multiplikativ hat NIX damit zu tun, ob der Datensatz stationär ist!!!!

(a) additative Modell
- man nutzt es wenn die saisonalen Schwankungen sind gleich groß egal wie hoch der Trend ist
- man nutzt es, wenn es gar keinen Trend gibt

(b) multiplikatives Modell
- man nutzt es wenn die saisonalen Schwankungen wachsen mit dem Trend (z.B. bei Umsätzen die Jahr für Jahr steigen)
- Multiplikativ setzt voraus dass alle Werte positiv sind – bei Nullwerten oder negativen Werten schlägt die multiplikative Zerlegung fehl oder liefert unsinnige Ergebnisse
- macht v.a. Sinn wenn Daten einen Trend haben. Ansonsten besser additiative Modell.

### Durchführung

```python
from statsmodels.tsa.seasonal import seasonal_decompose

# Set frequenc to daily
# This is done for seasonal decomposition to infere seasonal patterns
df = df.asfreq('D')

# Perform seasonal decomposition
decomposition = seasonal_decompose(df['unit_sales'], model='additive')

# Plotting
fig = decomposition.plot()
fig.set_size_inches(14, 6)  # (Breite, Höhe)
plt.tight_layout()
plt.show()
```

### Interpretation

- => Trend + seasonal + residual = oberste plot (Sprich der oberste Plot setzt sich additativ aus den 3 darunter zusammen)

- Trend:
    - Time Series schwankt zwischen 342 und 661
    - kein systematisches Steigen oder Fallen, sondern pendelt um einen stabilen Mittelwert
    - → kein Trend

- Saisonalität:
    - regelmäßiges, wöchentliches Muster

- Residual:
    - zufälliges Rauschen ohne erkennbares Muster → unauffällig

- Fazit:
    - wöchentliche Saisonalität ohne Trend

### ADF-Test
#### Allgemeines zum Test

- Anstieg/Abfall der zu untersuchender Größe aus Zeitreihe
- ADF Test guckt, ob anschließend dieser Ausschlag abklingt und Kennzahl auf langfristigen Mittelwert zurück kehrt (mean reversion)
    - Ausschlag klingt ab => Stationarität:
         - Es liegt Keine Einheitswurzel (auf die getestet wird) vor.
         - Das bedeutet anders gesagt, dass es war ein zufälliger Ausschlag und Daten sind stationär. Trotzdem kann Saisonalität enthalten sein, die die Einheitswurzel NICHT sieht.
               - => wichtig die Daten auch visuell zu überprüfen!!
    - Ausschlag klingt nicht ab => Keine Stationarität:
      - Es liegt eine Einheitswurzel (auf die getestet wird) vor.
      - Das bedeutet anders gesagt, dass keine stationären Daten vorliegen. Ob hinter dem Anstieg/Abfall ein Trend oder Saisonalität als Ursache liegt, wird nicht untersucht.

#### Durchführung

```python
from statsmodels.tsa.stattools import adfuller

# Perform Augmented Dickey-Fuller test
adf_result = adfuller(df['unit_sales'])
adf_result[1]
```

```python
from statsmodels.tsa.stattools import adfuller

# Perform Augmented Dickey-Fuller test
adf_result = adfuller(df['unit_sales'])
adf_result_dict = {
    'ADF Statistic': adf_result[0],
    'p-value': adf_result[1],
    'Critical Values': adf_result[4]
}
adf_result_dict
```

#### Interpretation

(a) ADF-Test: p-Wert
p < 0.05 → stationär

(b) ADF-Test: ADF-Statistik + Critical Values:
ADF-Statistik < Critical Values (1%, 5%, 10%) → stationär

(c) VORSICHT:
- Daten sind stationär im Sinne der Einheitswurzel. Es liegt also keine Einheitswurzel vor => Schocks klingen ab.
    - Saisonalität ist ein stabiles, vorhersehbares Muster – kein zufälliger Drift. => das sieht die Einheitswurzel NICHT.
- aber Daten sind nicht vollständig stationär wegen vorhandener Saisonalität (siehe visuelle Inspektion)

(d) Was bedeutet das:
- Modelle wie ARIMA setzen Stationarität im Sinne der Einheitswurzel voraus → direkt verwendbar
- Modelle wie SARIMA setzen zusätzlich Behandlung der Saisonalität voraus → besser geeignet

(e) abschließendes Urteil zur Stationarität
- Daten sind laut ADF/Einheitswurzel stationär (kein Drift) => ARIMA kann genutzt werden
- Daten sind allerdings laut visueller Inspektion nicht stationär, da wöchentliche Saisonalität vorhanden => SARIMA besser als ARIMA

## Smoothing Time Series
### Theorie

- man nutzt typischerweiße MEAN oder MEDIAN einer bestimmten Anzahl von Tagen / Jahren aus der Vergangenheit

........................
.. (1) MEAN-Smoothing ..
........................
- nimmt man 3 Zeiteinheiten für Mean-Berechnung => called "3 Mean Smoothing":
    - erste und letzte Wert bleibt frei



- also ergibt sich:



- Vorsicht vor Over-Smoothing. 3 oder 5 oft guter Startpunkt.

--------------------------------------------------------------------------------------------------------------------

- wenn statt ungerade eine gerade Zahl genommen wird -> wir müssen auch noch centern, da es keinen Mittelpunkt von 4 Werten gibt:
    - es also nach 4 Mean Smoothing nochmal 2 Mean Smooting von 2 Zahlen die sich durchs 4 Mean Smooting ergeben haben gemacht
    - => Bspsweise für 2004 ergibt sich 14.875 usw.




..........................
.. (2) MEDIAN-Smoothing ..
..........................


- bei geradzahliger MEDIAN-Smoothing muss man auch wieder centern





### Mean 7 Smoothing (genau: Lag-1 Rolling Average)

```python
# (1) 7-Mean Smooting
# = Lag-1 Rolling Average
# = verzögerter gleitender Durchschnitt

"""
(1) Shift 1
> damit ich den Wert vom aktuellen Tag nicht mit in Zeitreihe einbeziehe, sonst data leakage
> Rolling-7-Mean für Tag 10 (!ohne Shift) = Mittelwert aus Tag 4, 5, 6, 7, 8, 9, 10 => Tag 10 selbst drin
> Rolling-7-Mean für Tag 10 (mit Shift)   = Mittelwert aus Tag 3, 4, 5, 6, 7, 8, 9  => nur Vergangenheit
"""

time_series_day_category['rolling_mean_7'] = time_series_day_category['unit_sales'].shift(1).rolling(window=7).mean()

time_series_day_category
```

```python
# (2) Datenvorbereitung für Plotting

df = time_series_day_category.copy()

df_long = df.melt(
    id_vars='date',
    value_vars=['unit_sales', 'rolling_mean_7'],
    var_name='variante',
    value_name='wert',
)

print(df_long["variante"].unique())
print()
df_long.head(10)
```
